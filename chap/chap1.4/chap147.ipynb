```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('Elna.png',0)
edges = cv2.Canny(img,100,200)
edges1 = cv2.Canny(img,100,120)
```

```python
plt.figure(figsize=(15,20))
plt.subplot(131),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(edges1,cmap = 'gray')
plt.title('Edge Image2'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
edges1 = cv2.Canny(img,100,120)
blur = cv2.GaussianBlur(img,(3,3),0)
edges2 = cv2.Canny(blur,100,120)
plt.figure(figsize=(15,20))
plt.subplot(121),plt.imshow(edges1,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges2,cmap = 'gray')
plt.title('Edge Image with smooth'), plt.xticks([]), plt.yticks([])
plt.show()
```
